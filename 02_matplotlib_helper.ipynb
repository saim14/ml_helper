# Basics

**Example workflow**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#matplotlib object oriented approach
x = [1,2,3,4]
y = [2,4,7,9]

fig, ax = plt.subplots(figsize=(5,4))
ax.plot(x,y, color='orange')
ax.set(title = "Simple plot",
       xlabel = 'x-axis',
       ylabel = 'y-axis')
#fig.savefig('sample-plot.png')
plt.show()
```

**Multiple figure**

```python
x = [1,2,3,4]
data = {
    "saim": 80,
    "paim": 78,
    "daim": 60
}
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
ax1.plot(x,y, color='red');
ax2.scatter(np.random.random(10), np.random.random(10), color='g');
ax3.bar(data.keys(), data.values(), color='orange');
ax4.hist(np.random.randn(1000), color='black');
```

## Whiche one to use when? 
* When ploting something quickly, okay to use pyplot method, example above.
* When ploting something more advanced, use the OO methos. 

```python
# OO method, basic graph
fig, ax = plt.subplots(figsize=(10,6))
heart_df.plot(kind='scatter',
              x='age',
              y='chol',
              c='target',
              ax=ax);
```

**OO method full workflow**

```python
plt.style.use('seaborn-whitegrid')
# OO method from scratch 
fig, ax = plt.subplots(figsize=(10,6))
# plot the data
scatter = ax.scatter(x=heart_df['age'],
                     y=heart_df['chol'],
                     c=heart_df['target'],
                     cmap="winter")
# customize the plot
ax.set(title='Heart disease and cholesterol levels',
       xlabel='Age',
       ylabel='Cholesterol')
# customize for different graph
ax.set_xlim([30,80])
ax.set_ylim([100,450])

# add a lagend
ax.legend(*scatter.legend_elements(), title='Target');
# add a horizontal line for mean cholesterol
ax.axhline(heart_df['chol'].mean(),
           linestyle='--');
```

# More advaned graphs (multiple plot in single figure)

```python
# use matplotlib graph styles
plt.style.use('seaborn-whitegrid')
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)
# plot the data for axis 1 as ax0
scatter = ax0.scatter(x=heart_df['age'],
                      y=heart_df['chol'],
                      c=heart_df['target'],
                      cmap='winter')
# customize the plot
ax0.set(title='Heart disease and cholesterol levels',
       ylabel='Cholesterol')
ax0.set_ylim([100,500])
# add a lagend
ax0.legend(*scatter.legend_elements(), title='Target');
# add a horizontal line for mean cholesterol
ax0.axhline(heart_df['chol'].mean(),
           linestyle='--');

# plot the data for axis 2 as ax1
scatter = ax1.scatter(x=heart_df['age'],
                      y=heart_df['thalach'],
                      c=heart_df['target'],
                      cmap='winter')
# customize the plot
ax1.set(title='Heart disease and max heart rate',
        xlabel='Age',
        ylabel='Max heart rate')
ax1.set_xlim([30,80])

# add a lagend
ax1.legend(*scatter.legend_elements(), title='Target');
# add a horizontal line for mean cholesterol
ax1.axhline(heart_df['thalach'].mean(),
           linestyle='--');
# add a titile to figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
